````python
import numpy as np

def gradient_descent(
    gradient, start, learn_rate, n_iter=50, tolerance=1e-06
):
    vector = start
    for _ in range(n_iter):
        diff = -learn_rate * gradient(vector)
        #complete the if statement
        if np.all(np.abs(diff) <= tolerance):
            break
        vector += diff
    return vector
````

Now that you have defined the function, you can test the implementation of gradient descent by calling the `gradient_descent` *function*.

Note the following: You have set the `gradient` *argument* equal to $2x$ which represents the derivative of $f(x) = x^2$. The starting point, `start`, is set equal to 10 and the learning rate, `learn_rate`, is set equal to 0.2.

````python
gradient_descent(
    gradient=lambda v: 2 * v, start=0.4, learn_rate=0.2
)
````
